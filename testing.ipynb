```python
import pandas as pd
import datetime as dt
from icalendar import Calendar, Event
```

```python
schedule_name = "schedules/winter-miles.csv"
```

```python
schedule = pd.read_csv(schedule_name, header=None, skiprows=1, names=["description"])
```

```python
schedule.about
```

```python
schedule = schedule.reset_index()
with open(schedule_name) as f:
    schedule.about = f.readline()
```

```python
start_date = dt.date(year=2021, month=10, day=25)
schedule['date'] = schedule['index'].apply(lambda x: start_date + dt.timedelta(days=x))
```

```python
schedule.head()
```

```python
cal = Calendar()
cal.add('prodid', '-//Nithin Saripalli//Running Training Planner//EN')
cal.add('version', '2.0')
for index, row in schedule.iterrows():
    cal.add_component(Event({"summary": row.description, "DTSTART;VALUE=DATE": row.date.strftime("%Y%m%d"), "DTEND;VALUE=DATE": (row.date + dt.timedelta(days=1)).strftime("%Y%m%d"), "description": f"For details, click here {schedule.about}"}))
```

```python
cal.subcomponents
```

```python
cal.to_ical()
```

```python
import tempfile, os
with open('winter.ical', 'wb') as f:
    f.write(cal.to_ical())
    print(f.name)
```
